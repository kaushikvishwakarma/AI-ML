```python
x=[1,2,3,4]
y=[11,22,33,44]
plt.plot(x,y)
```

```python
# 1st method 
fig=plt.figure()# Create a figure
ax=fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method
fig =plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
#3rd method (recommended)
fig,ax=plt.subplots()
ax.plot(x,[50,100,200,250]);
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[11,22,33,44]
fig,ax=plt.subplots(figsize=(2,2))
ax.plot(x,y)
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")
#fig.savefig("C:\\Users\\vkaus\\OneDrive\\Documents\\ML\\images\\sample-plot.png")
```
