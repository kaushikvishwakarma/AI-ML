```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
            c='target',
            figsize=(10,6));
```

```python
## OO method from scratch
fig,ax=plt.subplots(figsize=(9,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50['target'])
# now customising 
ax.set(title='heart disease and Cholesterol levels',
    xlabel='Age of Patient',
       ylabel='Cholestrol level',
       );
ax.legend(*scatter.legend_elements(),title='Target');
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                           figsize=(10,6)
                          )
#add data
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50['chol'],
                   c=over_50['oldpeak'])
#customising ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        xlabel='Age',
       ylabel='Cholestrol');
ax0.axhline(over_50["chol"].mean(),linestyle=":");
ax0.legend(*scatter.legend_elements(),title="Target");


# Now for ax1
#add data
scatter1=ax1.scatter(x=over_50["age"],
                    y=over_50['thalach'],
                    c=over_50['target'])
ax1.set(title="Thalach",xlabel='Age',ylabel='Max heart rate')
ax1.legend(*scatter1.legend_elements(),title='target')
ax1.axhline(over_50["thalach"].mean(),linestyle=":");


#add title to figure 
fig.suptitle("Piyush ka Data",fontsize=16,fontweight="bold");
```
